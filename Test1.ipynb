```python
import pandas as pd
```

```python
df=pd.DataFrame({"Name":["anu","ravi","hari","sara"],
                 "Marks":[45,66,78,33]})
df
```

```python
df["Grade"]=df["Marks"].apply(lambda x:"A" if x>=70 else("B" if x>=50 else "c"))
print(df)
```

```python
print(df[df["Marks"]>=50])
```

```python
print(df["Marks"].mean())
print(df["Marks"].max())
```

```python
df=df.rename(columns={"Marks":"Score"})
df=df.drop("Grade",axis=1)
print(df)
```
