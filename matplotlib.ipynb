```python
import pandas as pd
```

```python
d=pd.read_csv("StudentsPerformance.csv")
d
```

```python
d.columns
```

```python
import matplotlib.pyplot as plt
```

```python
d.groupby("gender")[['math score', 'reading score',
       'writing score']].mean().plot(kind="bar")
plt.title("grnder vs score")
plt.xlabel("gender")
plt.ylabel("score")
plt.legend()
plt.show()
```

```python
c=["5 star","dairy milk","kitkat"]
s=[230,330,234]
a=plt.bar(c,s)
plt.bar_label(a)#label the value
plt.xlabel("choclote company")
plt.ylabel("sales")
plt.show()
```

```python
plt.hist(d["math score"],bins=20)#histogram chart is for distribution ,show the ranges,not give x and y axis
plt.title("distribution of the math score")
plt.xlabel("score")
plt.ylabel("frequency")
plt.show()

```

```python
#scatter plot
plt.scatter(d["math score"],d["reading score"])
plt.title(" math score vs reading scor")
plt.xlabel("math score")
plt.ylabel("reading scor")
plt.show()
```

```python
#line plot
avg_s=d[["math score",'reading score',"writing score"]].mean()
avg_s
```

```python
x=avg_s.index
y=avg_s.values
```

```python
plt.plot(x,y)
plt.title("maths vs reading")
plt.xlabel("math score")
plt.ylabel("reading score")
plt.show()
```

```python
#box plot
plt.boxplot([d["math score"],d["reading score"]])
plt.title("maths vs reading")
plt.xlabel("math score")
plt.ylabel("reading score")
plt.show()
```
